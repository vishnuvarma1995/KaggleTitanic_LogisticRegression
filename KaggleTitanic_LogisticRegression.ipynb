```python
import pandas as pd
train_data = pd.read_csv('train.csv')

# Titanic kaggle data, where in the Survived colum, 1 represents the person surviving and 0 means they died.
df = pd.DataFrame(train_data[['PassengerId','Survived','Pclass','Sex','Age','SibSp','Fare']])
df = df.dropna()
df 
```

```python
# Convert male and female label to 1 or 0
mapping_sex = {'male': 1, 'female': 0}
df = df.replace({'Sex': mapping_sex})
df
```

```python
from sklearn import linear_model

X = df[['Pclass','Sex','Age','SibSp','Fare']]
y = df[['Survived']]

lr = linear_model.LogisticRegression()
lr.fit(X,y)
# lr.predict(X)
```

```python
import pylab as pl
# Visualise histograms of each parameter in dataframe
df.hist()
pl.show()
```

```python
test_data = pd.read_csv('test.csv')
df_test = pd.DataFrame(test_data[['PassengerId','Pclass','Sex','Age','SibSp','Fare']])

# Convert male and female label to 1 or 0
mapping_sex_test = {'male': 1, 'female': 0}
df_test = df_test.replace({'Sex': mapping_sex_test})
# All datapoint needed, so can't get rid of rows with NaN. Replace all NaN with mean values of column
df_test = df_test.fillna(df_test.mean())

df_test
```

```python
Xtest = df_test[['Pclass','Sex','Age','SibSp','Fare']]
Prediction = lr.predict(Xtest) # Make prediction of test data from train data
df_test['Survived'] = Prediction # Add 'Survived' column
df_test
```

```python
# Output new csv file with solution
df_test[['PassengerId','Survived']].to_csv('LogitRegressionSolution.csv', index = False)
```
